## Scraping Billboard 

```python
# let's create the soup first
import bs4
from bs4 import BeautifulSoup 
import requests
import pandas as pd
```

```python
# url from the top 100 billboard
url = 'https://www.billboard.com/charts/hot-100'
```

```python
#download the html with get request
response = requests.get(url)
```

```python
response.status_code #200 means successful    #400 bed request
```

```python
html = response.content #this is the pure html
```

```python
soup = BeautifulSoup(response.content, 'html.parser')  #creating a soup to
```

```python
type(soup)
# all the website is stored into soup
```

```python
# select the css for the songs
scr = soup.select(".chart-element__information__song")
```

```python
# for loop to append the song name to a list
name_song = []

for song in scr:
    name_song.append(song.text)
```

```python
name_song = name_song[:100]
```

```python
artists = soup.select(".chart-element__information__artist")
artists
```

```python
# let's get the artists names
artist_list = []

for artist in artists:
    artist_list.append(artist.text)

artist_list
```

```python
len(name_song)
```

```python
# query the ranking
ranking = soup.select('.chart-element__rank__number')
```

```python
ranking
```

```python
#now we create teh list witht he ranking
rank_list = []

for rank in ranking:
    rank_list.append(rank.text)
    
rank_list
```

```python
# create the dataframe with the variables
df = pd.DataFrame({'Rank': rank_list,
                  'Artists': artist_list,
                   'Song': name_song })
df
```

```python
df.to_csv(r'Billboard_scraping.csv', index = False, header=True)
```
